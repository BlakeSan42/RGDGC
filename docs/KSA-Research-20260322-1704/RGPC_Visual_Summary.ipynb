# River Grove Parks Conservancy — Visual Summary

### Shareable charts & data for quick understanding

**What this is:** A community-driven initiative to modernize River Grove Park in Kingwood, TX with technology, transparency, youth employment, and community programming — at zero additional cost to residents.

**The one-liner:** *"We're not trying to take over. We're trying to bring Kingwood's parks into the 21st century — with technology, transparency, and jobs for our kids."*

```python
import matplotlib
matplotlib.use('Agg')
import matplotlib.pyplot as plt
import matplotlib.ticker as mticker
import numpy as np
import json, os

plt.rcParams.update({
    'figure.facecolor': 'white', 'axes.facecolor': '#f8f9fa',
    'font.family': 'sans-serif', 'font.size': 11,
    'axes.grid': True, 'grid.alpha': 0.3,
    'figure.dpi': 150, 'savefig.dpi': 150,
})

FOREST = '#1B5E20'; ORANGE = '#FF6B35'; RED = '#B71C1C'; BLUE = '#1565C0'
GRAY = '#424242'; LIGHT_GREEN = '#E8F5E9'; PURPLE = '#7B1FA2'

OUT = '/Users/blakesanders/RGDGC/docs/KSA-Research-20260322-1704/charts'
os.makedirs(OUT, exist_ok=True)

print("Setup complete — generating charts...")
```

## 1. KSA Historical Finances — 14 Years of 990 Data

```python
# KSA 990 Data (2011-2024)
years = list(range(2011, 2025))
revenue = [819888,662415,845378,760477,838333,904564,884034,926473,998251,1031234,1039957,1034863,1041801,1048921]
expenses = [743479,724047,911275,649890,771855,789390,895731,745644,850536,1018430,948240,861519,1009164,1565650]
surplus = [r - e for r, e in zip(revenue, expenses)]
net_assets = [2913747,2729617,2669904,2780491,2846971,2962145,2950448,3196991,3344706,3357510,3449227,3604370,3637007,3120278]

fig, axes = plt.subplots(2, 2, figsize=(14, 10))
fig.suptitle('KSA Financial History (2011-2024) — IRS Form 990 Data', fontsize=16, fontweight='bold', color=FOREST)

# Revenue vs Expenses
ax = axes[0, 0]
x = np.arange(len(years))
w = 0.35
ax.bar(x - w/2, [r/1e6 for r in revenue], w, label='Revenue', color=FOREST, alpha=0.85)
ax.bar(x + w/2, [e/1e6 for e in expenses], w, label='Expenses', color=ORANGE, alpha=0.85)
ax.bar(x[-1] + w/2, expenses[-1]/1e6, w, color=RED, alpha=0.9)  # Highlight 2024 spike
ax.set_xticks(x); ax.set_xticklabels(years, rotation=45, fontsize=8)
ax.set_ylabel('Millions ($)'); ax.set_title('Revenue vs Expenses', fontweight='bold')
ax.legend(fontsize=9); ax.set_ylim(0, 1.8)
ax.annotate('55% SPIKE\n$1.57M', xy=(13 + w/2, 1.57), fontsize=8, ha='center', va='bottom', color=RED, fontweight='bold')

# Surplus / Deficit
ax = axes[0, 1]
colors = [FOREST if s > 0 else RED for s in surplus]
ax.bar(years, [s/1e3 for s in surplus], color=colors, alpha=0.85)
ax.axhline(y=0, color=GRAY, linewidth=0.8)
ax.set_ylabel('Thousands ($)'); ax.set_title('Annual Surplus / (Deficit)', fontweight='bold')
ax.set_xticklabels(years, rotation=45, fontsize=8)
ax.annotate('-$517K', xy=(2024, -517), fontsize=9, ha='center', va='top', color=RED, fontweight='bold')

# Net Assets
ax = axes[1, 0]
ax.fill_between(years, [n/1e6 for n in net_assets], alpha=0.3, color=FOREST)
ax.plot(years, [n/1e6 for n in net_assets], color=FOREST, linewidth=2.5, marker='o', markersize=4)
ax.set_ylabel('Millions ($)'); ax.set_title('Net Assets (Should Decline if Surplus Returned)', fontweight='bold')
ax.set_xticklabels(years, rotation=45, fontsize=8)
ax.annotate('Peak: $3.64M\n(Article VIII says\nreturn surplus)', xy=(2023, 3.64), fontsize=8,
            ha='center', va='bottom', color=RED, fontweight='bold',
            arrowprops=dict(arrowstyle='->', color=RED))

# Cumulative Retained Surplus
cumul = []
running = 0
for s in surplus:
    if s > 0:
        running += s
    cumul.append(running)
ax = axes[1, 1]
ax.fill_between(years, [c/1e6 for c in cumul], alpha=0.3, color=RED)
ax.plot(years, [c/1e6 for c in cumul], color=RED, linewidth=2.5, marker='s', markersize=4)
ax.set_ylabel('Millions ($)'); ax.set_title('Cumulative Retained Surplus (Owed to Villages)', fontweight='bold')
ax.set_xticklabels(years, rotation=45, fontsize=8)
ax.annotate(f'$1.01M\nretained', xy=(2023, cumul[-2]/1e6), fontsize=9, ha='right', va='bottom', color=RED, fontweight='bold')

plt.tight_layout()
plt.savefig(f'{OUT}/01_ksa_financials.png', bbox_inches='tight')
plt.show()
print("Chart 1 saved: 01_ksa_financials.png")
```

## 2. KSA vs RGPC — The Comparison That Matters

```python
# KSA vs RGPC comparison
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 6))
fig.suptitle('KSA (Current) vs RGPC (Proposed) — Same $41/unit/year', fontsize=15, fontweight='bold', color=FOREST)

# Bar comparison
categories = ['Youth\nEmployed', 'Annual\nEvents', 'Financial\nReports/Yr', 'Complaint\nResolution %', 'Flood\nSensors', 'App\nDownloads']
ksa_values = [0, 0, 0, 0, 0, 0]
rgpc_values = [60, 25, 12, 82, 5, 2700]
maxvals = [60, 25, 12, 100, 5, 2700]
ksa_norm = [k/m*100 for k, m in zip(ksa_values, maxvals)]
rgpc_norm = [r/m*100 for r, m in zip(rgpc_values, maxvals)]

x = np.arange(len(categories))
w = 0.35
bars1 = ax1.bar(x - w/2, ksa_norm, w, label='KSA (Current)', color=RED, alpha=0.7)
bars2 = ax1.bar(x + w/2, rgpc_norm, w, label='RGPC (Projected)', color=FOREST, alpha=0.85)
ax1.set_xticks(x); ax1.set_xticklabels(categories, fontsize=9)
ax1.set_ylabel('% of Potential (normalized)')
ax1.set_title('Service Delivery Comparison', fontweight='bold')
ax1.legend()
ax1.set_ylim(0, 115)

for bar, val in zip(bars1, ksa_values):
    ax1.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 2, str(val),
             ha='center', va='bottom', fontsize=9, color=RED, fontweight='bold')
for i, (bar, val) in enumerate(zip(bars2, rgpc_values)):
    label = f'{val:,}' if val >= 100 else (f'{val}%' if i == 3 else str(val))
    ax1.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 2, label,
             ha='center', va='bottom', fontsize=9, color=FOREST, fontweight='bold')

# Revenue projection
proj_years = ['Year 1', 'Year 2', 'Year 3', 'Year 4', 'Year 5']
proj_rev = [100, 406, 616, 720, 830]
proj_exp = [178, 382, 575, 640, 710]

ax2.fill_between(range(5), proj_rev, alpha=0.3, color=FOREST)
ax2.plot(range(5), proj_rev, color=FOREST, linewidth=2.5, marker='o', label='Revenue', markersize=6)
ax2.plot(range(5), proj_exp, color=ORANGE, linewidth=2.5, marker='s', label='Expenses', markersize=6)
ax2.fill_between(range(5), proj_rev, proj_exp, where=[r > e for r, e in zip(proj_rev, proj_exp)],
                 alpha=0.2, color=FOREST, label='Surplus')
ax2.fill_between(range(5), proj_rev, proj_exp, where=[r <= e for r, e in zip(proj_rev, proj_exp)],
                 alpha=0.2, color=RED, label='Gap (Y1)')
ax2.set_xticks(range(5)); ax2.set_xticklabels(proj_years)
ax2.set_ylabel('Thousands ($K)')
ax2.set_title('RGPC Revenue vs Expenses — Self-Sustaining by Year 2', fontweight='bold')
ax2.legend(fontsize=9)
ax2.annotate('Self-sustaining\nfrom here →', xy=(1.5, 400), fontsize=9, color=FOREST, fontweight='bold')

plt.tight_layout()
plt.savefig(f'{OUT}/02_ksa_vs_rgpc.png', bbox_inches='tight')
plt.show()
print("Chart 2 saved: 02_ksa_vs_rgpc.png")
```

## 3. Social Impact Monte Carlo — 1,000 Simulations

```python
# Load Monte Carlo social impact data
with open('/Users/blakesanders/RGDGC/docs/KSA-Research-20260322-1704/social_impact_raw_data.json') as f:
    mc = json.load(f)

scores = [d['score'] for d in mc]
youth = [d['youth_5yr'] for d in mc]
wages = [d['wages_5yr'] for d in mc]
engaged = [d['engaged'] for d in mc]
statuses = [d['status'] for d in mc]
ksa_responses = [d['ksa'] for d in mc]
trees = [d['trees'] for d in mc]
trust = [d['trust'] for d in mc]

fig, axes = plt.subplots(2, 3, figsize=(16, 10))
fig.suptitle('Monte Carlo Social Impact — 1,000 Simulated Futures for RGPC', fontsize=16, fontweight='bold', color=FOREST)

# 1. Initiative outcome distribution
ax = axes[0, 0]
status_counts = {s: statuses.count(s) for s in ['thriving', 'stable', 'struggling', 'failed']}
colors_s = [FOREST, BLUE, ORANGE, RED]
bars = ax.barh(list(status_counts.keys()), list(status_counts.values()), color=colors_s, alpha=0.85)
ax.set_xlabel('Scenarios (out of 1,000)')
ax.set_title('Initiative Outcome', fontweight='bold')
for bar, val in zip(bars, status_counts.values()):
    ax.text(bar.get_width() + 10, bar.get_y() + bar.get_height()/2, f'{val} ({val/10:.1f}%)',
            va='center', fontsize=10, fontweight='bold')
ax.set_xlim(0, 600)

# 2. Social impact score histogram
ax = axes[0, 1]
ax.hist(scores, bins=20, color=FOREST, alpha=0.8, edgecolor='white')
median_score = np.median(scores)
ax.axvline(median_score, color=ORANGE, linewidth=2, linestyle='--', label=f'Median: {median_score:.1f}')
ax.set_xlabel('Social Impact Score (0-100)')
ax.set_ylabel('Scenarios')
ax.set_title('Social Impact Score Distribution', fontweight='bold')
ax.legend(fontsize=10)

# 3. Youth employed histogram
ax = axes[0, 2]
ax.hist(youth, bins=25, color=BLUE, alpha=0.8, edgecolor='white')
median_youth = np.median(youth)
ax.axvline(median_youth, color=ORANGE, linewidth=2, linestyle='--', label=f'Median: {median_youth:.0f} youth')
ax.axvline(0, color=RED, linewidth=3, label='KSA current: 0')
ax.set_xlabel('Unique Youth Employed (5yr)')
ax.set_ylabel('Scenarios')
ax.set_title('Youth Employment Distribution', fontweight='bold')
ax.legend(fontsize=9)

# 4. Wages paid histogram
ax = axes[1, 0]
wages_k = [w/1000 for w in wages]
ax.hist(wages_k, bins=25, color=PURPLE, alpha=0.8, edgecolor='white')
median_wage = np.median(wages_k)
ax.axvline(median_wage, color=ORANGE, linewidth=2, linestyle='--', label=f'Median: ${median_wage:.0f}K')
ax.set_xlabel('Total Wages Paid ($K, 5yr)')
ax.set_ylabel('Scenarios')
ax.set_title('Youth Wages Distribution', fontweight='bold')
ax.legend(fontsize=10)

# 5. Community trust score vs KSA
ax = axes[1, 1]
ax.hist(trust, bins=20, color=FOREST, alpha=0.8, edgecolor='white')
ax.axvline(np.median(trust), color=ORANGE, linewidth=2, linestyle='--', label=f'RGPC Median: {np.median(trust):.1f}')
ax.axvline(0, color=RED, linewidth=3, label='KSA: BBB F = 0')
ax.set_xlabel('Community Trust Score (0-10)')
ax.set_ylabel('Scenarios')
ax.set_title('Trust Score: RGPC vs KSA', fontweight='bold')
ax.legend(fontsize=9)

# 6. Impact by KSA response
ax = axes[1, 2]
for resp, color, label in [('cooperate', FOREST, 'KSA Cooperates'),
                             ('ignore', BLUE, 'KSA Ignores'),
                             ('resist', RED, 'KSA Resists')]:
    subset = [d['score'] for d in mc if d['ksa'] == resp]
    ax.hist(subset, bins=15, alpha=0.5, color=color, label=f'{label} (n={len(subset)})', edgecolor='white')
ax.set_xlabel('Social Impact Score')
ax.set_ylabel('Scenarios')
ax.set_title('Impact by KSA Response', fontweight='bold')
ax.legend(fontsize=9)

plt.tight_layout()
plt.savefig(f'{OUT}/03_monte_carlo_social.png', bbox_inches='tight')
plt.show()
print("Chart 3 saved: 03_monte_carlo_social.png")
```
